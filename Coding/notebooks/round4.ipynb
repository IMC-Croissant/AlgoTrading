```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_trade = pd.read_csv("D:\Florida State University\Spring 23\Security Society\\r4_prices_d1.csv", sep= ';')
```

```python

df_banana = csv_trade[csv_trade['product'] == 'BANANAS'].sort_values(by=["timestamp"])
df_pearl = csv_trade[csv_trade['product'] == 'PEARLS'].sort_values(by=["timestamp"])
df_pina_coladas = csv_trade[csv_trade['product'] == 'PINA_COLADAS'].sort_values(by=["timestamp"])
df_coconuts = csv_trade[csv_trade['product'] == 'COCONUTS'].sort_values(by=["timestamp"])
df_berries = csv_trade[csv_trade['product'] == 'BERRIES'].sort_values(by=["timestamp"])
df_bag = csv_trade[csv_trade['product'] == 'BAGUETTE'].sort_values(by=["timestamp"])
df_picnic = csv_trade[csv_trade['product'] == 'PICNIC_BASKET'].sort_values(by=["timestamp"])
df_dip = csv_trade[csv_trade['product'] == 'DIP'].sort_values(by=["timestamp"])
df_uku = csv_trade[csv_trade['product'] == 'UKULELE'].sort_values(by=["timestamp"])
df_dolphin = csv_trade[csv_trade['product'] == 'DOLPHIN_SIGHTINGS'].sort_values(by=["timestamp"])
df_diving_gear = csv_trade[csv_trade['product'] == 'DIVING_GEAR'].sort_values(by=["timestamp"])
```

```python
df_underlying = df_dip.set_index('timestamp').mid_price*4 + df_uku.set_index('timestamp').mid_price + df_bag.set_index('timestamp').mid_price*2
```

```python
df_picnic.set_index('timestamp', inplace= True)
```

```python
plt.plot(df_picnic.mid_price)
plt.plot(df_underlying)
plt.legend(['picnic', 'underlying'])
```

```python
df_spread = df_picnic.mid_price - df_underlying
```

```python
plt.plot(df_spread)
plt.axhline(df_spread.mean())
```
